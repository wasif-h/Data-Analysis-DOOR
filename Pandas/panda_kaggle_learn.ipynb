## Indexing, Selecting & Assigning


```python
df = football.head(10)
df
```

```python
df.Host
# remember without loc,iloc - everythings works as column x row
```

```python
# iloc - works on index, no lebel, can assign, slice works like range - 0:2 means 0,1
display(df.head(5))
print(df.iloc[0], end='\n\n')
print(df.iloc[0, 0], end='\n\n')
print(df.iloc[0, 1], end='\n\n')
print(df.iloc[0:3, 0:2], end='\n\n')
print(df.iloc[0:3, -2:], end='\n\n')
display(df.iloc[[0, 1, 2], :])
```

```python
# loc - works on lebeling/string, no index but if lebel is integer than it can grab it
# can assign, slice doesn't works like range - 0:2 means -> 0 1 2
df = football
display(df.head(3))
df.loc[0, 'TopScorrer']
df.index = df['Champion']
display(df.head(5))
```
